```python
#Here we calculate how good our approximation of ùê∏(ùëë(ùë•))~ùê∏(ùõ•ùëë(ùë• + 1)) is as a function of ùê∏(i) as describe on the Appendix
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
Ei = 10**(np.linspace(start=1, stop=2.5, num=100))
limit_sum = np.floor(Ei*(1-(4.9/np.sqrt(Ei))))
limit_sum[limit_sum < 0] = 0
limit_sum = limit_sum.astype(int)

exponential_terms = np.exp(-Ei)

SumProb = np.zeros(len(Ei))
SumiProb = np.zeros(len(Ei))

for i in range(len(Ei)):
    for j in range(limit_sum[i]):
        SumProb[i] = SumProb[i] + Ei[i]*exponential_terms[i]/math.factorial(i)
        SumiProb[i] = SumiProb[i] + i*Ei[i]*exponential_terms[i]/math.factorial(i)

Res = -limit_sum*SumProb + SumiProb

Approx = 4.9*np.sqrt(Ei) + 1

PerErr = 100*np.abs(Res)/(Res+Approx)
```

```python
plt.plot(Ei, Approx)
plt.xlabel('E(i)')
plt.ylabel('')
```

```python
plt.plot(Ei, PerErr)
plt.xlabel('E(i)')
plt.ylabel('')
```
